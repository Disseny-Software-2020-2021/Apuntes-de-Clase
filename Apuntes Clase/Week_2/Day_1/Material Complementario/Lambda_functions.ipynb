### Lambda functions

The lambda tends to intimidate people more than it should. This reaction seems to stem from the name “lambda” itself—a name that comes from the Lisp language, which got it from lambda calculus, which is a form of symbolic logic. 

It's an **anonymous function**. It's an expression that creates a function, but does not assign it to any name.

It can only contain a single expression, not a block of statements.

The syntax is:

*lambda inputs_separated_by_colon : value_to_return*


```python
f = lambda x, y, z: x + y + z
```

```python
f
```

Here in this example, the **inputs** are:

* x
* y
* z

and the value returned is: x + y + z

The type of the arguments is not fixed, the input variables can be of any type and is the programmer responsability to know hoy to combine the inputs to return a consisten value ( you can't add a number and a string ).

Lambda functions doesn't have a fixed number of inputs, they can have as many as desired. However, the number of inputs becomes fixed once the function is defined. Let's see another example

```python
z = lambda x: x + 1
```

```python
z(2)
```

```python
z(2,3)
```

Why you should care about lambda functions? They are cool to do some quick calculations

```python
my_list = [1,2,3,4,5]

for elem in my_list:
    print(z(elem))
```
